```python
# create 100 random phase angles [0,2pi] and show unit vectors with those angles

n = 100

randphases = np.random.rand(n)*2*np.pi

for i in range(0,n):
    plt.polar([0,randphases[i]],[0,1],color=np.random.rand(3),markersize=20,alpha=.3)

plt.show()
```

```python
# create an outwards spiral using phase angles and amplitudes

n = 100
a = np.linspace(0,1,n)
p = np.linspace(0,4*np.pi,n)
# x = a * np.cos(p)
# y = a * np.sin(p)
# plt.plot(x,y);
plt.polar(p,a,'m')
plt.axis('off')
plt.show()
```

```python
# convert radians to degrees

n = 10
rad = np.logspace(np.log10(.001),np.log10(2*np.pi),n)

print(np.rad2deg(rad))
```

```python
# famous equality in trigonometry

ang = np.logspace(np.log10(1),np.log10(2*np.pi),10)
r = np.cos(ang)*2 + np.sin(ang)*2
print(r)
```

```python
# create euler's number
p = np.pi/4
m = .5

eulr = m*np.exp(1j*p)

# now extract magnitude and phase
mag = np.abs(eulr)
ang = np.angle(eulr)

# then plot
plt.polar([0,ang],[0,mag],'b',linewidth=3)
plt.show()
```
